```python
with open('data/names.txt') as fobj:
    names = fobj.read().split('\n') # split file context at newline
# map letters to integer indices, including '.' (end-of-word character)
# this will help us build a transition matrix
alphabet = ".abcdefghijklmnopqrstuvwxyz"
idx = {letter:index for index, letter in enumerate(alphabet)}
idx

all_names = '.'+'.'.join(names)
```

```python
# tally up all letter transitions

import numpy as np
transition_counts = np.zeros((27, 27*27))
for char1, char2, char3 in zip(all_names[:-2], all_names[1:-1], all_names[2:]):
    # print(char1, char2, char3)
    # what should the order of indices be?
    transition_counts[idx[char3], 27*idx[char1] + idx[char2]] += 1
```

```python
# normalize to get transition probabilities
transtion_probabilities = transition_counts.copy()

for i in range(27*27):
    # normalize the i-th column
    transtion_probabilities[:, i] /= transtion_probabilities[:, i].sum()
```

```python
# generate words
import numpy as np

word = '.a' # need to start with two letters to predict the third

def generate_word(starting_pair = ".a"):
    word = starting_pair
    done = False

    rng = np.random.default_rng()

    while not done:
        prev_char, curr_char = word[-2:]
        i1 = idx[prev_char]
        i2 = idx[curr_char]
        column = 27*i1 + i2

        p = transtion_probabilities[:, column]
        if any(np.isnan(p)):
            j = rng.choice(27)
        else:
            j = rng.choice(27, p=transtion_probabilities[:, column])
        next_char = alphabet[j]
        word = word+next_char

        if next_char=='.':
            done=True
    return word
```

```python
for _ in range(10):
    print(generate_word('..'))
```
